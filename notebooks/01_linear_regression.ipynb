# Regresión Lineal
## ¿Qué es la Regresión Lineal?

La regresión lineal es uno de los algoritmos más fundamentales en machine learning. Su objetivo es encontrar la mejor línea recta que se ajuste a un conjunto de datos, permitiéndonos hacer predicciones sobre nuevos valores.

### Conceptos Clave:

**Modelo Matemático:** y = mx + b
- **m**: pendiente de la línea (slope)
- **b**: intercepto con el eje y (y-intercept)
- **x**: variable independiente (feature)
- **y**: variable dependiente (target)

### ¿Cómo funciona?

1. **Datos**: Tenemos puntos (x, y) dispersos
2. **Objetivo**: Encontrar la línea que mejor se ajuste a estos puntos
3. **Método**: Minimizar la suma de errores cuadráticos (least squares)

### Aplicaciones:
- Predecir precios de casas basado en su tamaño
- Estimar ventas futuras basado en inversión publicitaria
- Analizar relaciones entre variables


```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Configuración de datos sintéticos
size = 100  # Número de puntos a generar
bot_limit, top_limit = 0, 100  # Límites para los valores de X

# Generamos datos aleatorios para X
X = np.random.randint(bot_limit, top_limit, size=size)

# Parámetros reales del modelo (los que queremos "descubrir")
m_real = np.random.rand() * np.random.choice([1, -1])  # Pendiente real aleatoria entre 0 y 1 o 0 y -1
b_real = np.random.randint(0, 10)  # Intercepto real aleatorio entre 0 y 10

# Agregamos ruido a nuestros datos (simulando datos del mundo real)
epsilon = np.random.normal(loc=0, scale=6.0, size=size)

# Función real que genera nuestros datos
f_real = lambda x: m_real * x + b_real + epsilon
y = f_real(X)
```

```python
# Configuración de colores para las gráficas
bg_color = '#' + '#F9F6F3'.strip('#')  # Color de fondo
accent = '#' + '437057'.strip('#')      # Color de acento

# Crear la primera gráfica: datos originales
fig, ax = plt.subplots(facecolor=bg_color)

# Scatter plot de los datos originales
ax.scatter(X, y, alpha=0.8, c=accent)

# Configuración estética de la gráfica
ax.set_facecolor(bg_color)
ax.set_xticklabels([])  # Removemos etiquetas del eje x
ax.set_yticklabels([])  # Removemos etiquetas del eje y
ax.spines['top'].set_visible(False)     # Removemos borde superior
ax.spines['right'].set_visible(False)   # Removemos borde derecho
plt.title('Datos Originales - Dispersión')
plt.show()
```

# REGRESIÓN LINEAL MANUAL (implementación desde cero)

```python
# Calculamos las medias de X e y
x_mean = np.mean(X)
y_mean = np.mean(y)

# Fórmula para calcular la pendiente (slope)
# m = Σ[(xi - x̄)(yi - ȳ)] / Σ[(xi - x̄)²]
m_hat = np.sum((X - x_mean) * (y - y_mean)) / np.sum((X - x_mean) ** 2)

# Fórmula para calcular el intercepto
# b = ȳ - m * x̄
b_hat = y_mean - m_hat * x_mean

# Función de predicción con nuestros parámetros calculados
f_hat = lambda x: m_hat * x + b_hat
y_hat = f_hat(X)
```

```python
# Creamos puntos para dibujar la línea de regresión
x_model = np.linspace(bot_limit, top_limit, num=1000)
y_model = f_hat(x_model)
```

```python
# Configuración de colores para la segunda gráfica
bg_color = '#' + '#F9F6F3'.strip('#')
accent = '#' + 'B9375D'.strip('#')      # Color para la línea de regresión
accent2 = '#' + '437057'.strip('#')     # Color para los datos originales

# Crear la segunda gráfica: datos + línea de regresión
fig, ax = plt.subplots(facecolor=bg_color)

# Datos originales
ax.scatter(X, y, alpha=0.8, c=accent2, label='Datos originales')
# Línea de regresión
ax.plot(x_model, y_model, c=accent, linewidth=2, label='Regresión lineal')

# Configuración estética
ax.set_facecolor(bg_color)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.title('Regresión Lineal - Implementación Manual')
plt.legend()
plt.show()
```

# COMPARACIÓN CON SKLEARN

```python
from sklearn.linear_model import LinearRegression
```

```python
# Creamos y entrenamos el modelo de sklearn
model = LinearRegression()
# Nota: sklearn requiere que X sea una matriz 2D, por eso usamos reshape(-1, 1)
model.fit(X.reshape(-1, 1), y)

# Extraemos los parámetros calculados por sklearn
m_model = model.coef_[0]      # Pendiente
b_model = model.intercept_    # Intercepto
```

```python
print("=== COMPARACIÓN DE RESULTADOS ===")
print(f"Pendiente real    = {m_real:.3f}")
print(f"Pendiente sklearn = {m_model:.3f}")
print(f"Pendiente manual  = {m_hat:.3f}")
print()
print(f"Intercepto real    = {b_real:.3f}")
print(f"Intercepto sklearn = {b_model:.3f}")
print(f"Intercepto manual  = {b_hat:.3f}")

print("\n=== ANÁLISIS ===")
print("- Nuestra implementación manual debe dar resultados muy similares a sklearn")
print("- Los valores 'reales' incluyen ruido, por lo que es normal que difieran")
print("- El objetivo es que nuestros métodos se acerquen lo más posible a los valores reales")
```
